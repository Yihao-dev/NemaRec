```python
import pandas as pd
import seaborn as sns
```

```python
tips=pd.read_csv('Desktop/scipy-2019-pandas-master/data/tips.csv')
tips.head()
```

```python
import matplotlib.pyplot as plt
tips.tip.plot(kind='hist')
```

```python
s=tips.smoker.value_counts()
print(s)
```

```python
s.plot(kind='bar')
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
sns.displot(tips.total_bill,kde=True)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='smoker',row='day')
#use hue to tease out the different group
#use col to tease out the smokers and non-smokers
#row and col are to create different variables etc
```

```python
facet=sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
facet.map(plt.scatter,'total_bill','tip')
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
ax1.scatter(tips.tip,tips.total_bill)
ax2.hist(tips.total_bill)
#subplot the histogram and scatter plot of the total bill and tips
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
sns.displot(tips.tip,kde=True,ax=ax1)
sns.regplot(x='total_bill',y='tip',data=tips,ax=ax2)
```

```python
sns.boxplot(x='sex',y='tip',data=tips)
```

```python
sns.boxplot(x='smoker',y='tip',data=tips)
```
