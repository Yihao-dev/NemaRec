```python
import pandas as pd
df=pd.read_csv('./Desktop/data.csv',header=None)
df
```

```python
#assign 30 features to the numpy array
from sklearn.preprocessing import LabelEncoder
X=df.iloc[:,2:].values#the features terms
y=df.iloc[:,1].values#the results
#encode the categorical data into the numerical values with LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)#encode and binarize the diagnosis results
le.transform(['M',"B"])
```

```python
#split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#combine the transformers and pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

#similar to the zip function, pipeline chains standardization, dimensionality and classifier together
pipe_lr=Pipeline([('scl',StandardScaler()),('pca',PCA(n_components=2)),('clf',LogisticRegression(random_state=1))])
#fit the data, much simpler than the traditional way where they are seperate
pipe_lr.fit(X_train,y_train)
```

```python
import pickle
# Save the Modle to file in the current working directory
#then drag it out to the desktop to facilitate future deployment
Pkl_Filename = "./Desktop/Pickle_EEG_Model.pkl"  

with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(pipe_lr, file)
```

```python
with open(Pkl_Filename, 'rb') as file:  
    Model = pickle.load(file)

Model
```

```python
score = Model.score(X_train, y_train)  
# Print the Score
print("Test score: {0:.2f} %".format(100 * score))  

# Predict the Labels using the reloaded Model
Ypredict = Model.predict(X_train)  
Ypredict
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy',max_depth=None)
from sklearn.metrics import accuracy_score

tree=tree.fit(X_train,y_train)
y_train_pred=tree.predict(X_train)
#y_test_pred=tree.predict(X_test)
```
